```python
# importation des librairies
import pandas as pd
import datetime
from sklearn.preprocessing import OneHotEncoder
```

```python
# importation des données
df  = pd.read_csv("voitures.csv",sep=';')
df_voitures = df.copy()
df_voitures.head()
```

```python
# suppression de colonne matricule
df_voitures = df_voitures.drop(columns=['matricule'], axis=1)
df_voitures.head()
```

```python
# transformation de la colonne date_circulation en année
df_voitures['date_ciruclation'] = pd.to_datetime(df_voitures['date_ciruclation'], format='%d/%m/%Y')
df_voitures['annee'] = df_voitures['date_ciruclation'].dt.year
df_voitures
```

```python
# transformation de la colonne date_circulation en age
df_voitures['date_ciruclation'] = pd.to_datetime(df_voitures['date_ciruclation'], format='%d/%m/%Y')
df_voitures['age'] = datetime.datetime.now().year - df_voitures['date_ciruclation'].dt.year
df_voitures
```

```python
# suppression de la colonne date_circulation
df_voitures = df_voitures.drop(columns=['date_ciruclation'], axis=1)
df_voitures
```

```python
# transformation de la colonne marque en one hot encoding
OHE = OneHotEncoder()
transformed = OHE.fit_transform(df_voitures[['marque']])
df_voitures.drop(columns = ['marque'],inplace=True)
df_voitures["Marque_"+OHE.categories_[0]] = transformed.toarray()
df_voitures
```

```python
# transformation de la colonne couleur en one hot encoding
OHE = OneHotEncoder()
transformed = OHE.fit_transform(df_voitures[['couleur']])
df_voitures.drop(columns = ['couleur'],inplace=True)
df_voitures["Couleur"+OHE.categories_[0]] = transformed.toarray()
df_voitures
```
